```python
# 4.
car_sales = pd.read_csv('car-sales.csv')
```

```python
# 5.
car_sales.head()
```

```python
car_sales.info()
```

```python
car_sales['Price']=car_sales['Price'].str.replace(r'[/$,]', '',regex=True)
```

```python
car_sales['Price']=car_sales['Price'].astype('int64')
```

```python
#5.
car_sales.head()
```

```python
#6. 
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head()
```

```python
today=pd.Timestamp.now().normalize()
```

```python
car_sales['Sale Date']=pd.date_range(start=today, periods=len(car_sales))
```

```python
#7.
car_sales.head()
```

```python
# 8. Line plot of price and total sales
plt.plot(car_sales['Price'],
        car_sales['Total Sales'])
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs the total sales line plot')
plt.show()
```

```python
# 9. scatterplot of odometer and price columns
plt.scatter(car_sales['Odometer (KM)'],
        car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer vs Price Scatter Plot')
plt.show()
```

```python
# 10. bar graph of make and odometer
plt.bar(car_sales['Make'],
        car_sales['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make Vs Odometer Bar Plot')
plt.show()
```

```python
# 11. histogram of the make column
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title('Histogram of Car Make')
plt.show()
```

```python
# 12. the creation of a subplot of multiple plots
fig=plt.figure(figsize=(15,7))
ax1=fig.add_subplot(221)
ax1.set_title("Price vs Total Sales Line Plot")
ax1.plot(car_sales['Price'],
        car_sales['Total Sales'],
        color="blue")
ax2=fig.add_subplot(222)
ax2.set_title("Odometer vs Price Scatter Plot")
ax2.scatter(car_sales['Odometer (KM)'],
        car_sales['Price'],
        color="blue")
ax3=fig.add_subplot(223)
ax3.set_title("Make vs Odomater Bar Plot")
ax3.bar(car_sales['Make'],
        car_sales['Odometer (KM)'],
        color="blue")
ax4=fig.add_subplot(224)
ax4.set_title("Histogram of Car Make")
ax4.hist(car_sales['Make'],
        color="blue")
plt.show()
```

```python
# 13. box plot of colour and total sales, with the help of seaborn plot
sns.boxplot(x='Colour', y='Total Sales', data=car_sales)
plt.xlabel('Colour')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 14. using linear regression to find out the relationship of total sales and odometer
sns.lmplot(x='Total Sales',
           y='Odometer (KM)',
           data=car_sales,
           hue='Make',
          ci=None)
plt.xlabel('Total Sales')
plt.show()
```

```python
# 15. pair plot
sns.pairplot(car_sales)
plt.show()
```

```python
#16. correlation between total sales, price, and odometer
correlation_matrix = car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
sales_price_corr = correlation_matrix.loc['Total Sales', 'Price']
sales_odometer_corr = correlation_matrix.loc['Total Sales', 'Odometer (KM)']
price_odometer_corr = correlation_matrix.loc['Price', 'Odometer (KM)']
print("\nPearson Correlation between Total Sales and Price:", sales_price_corr)
print("Pearson Correlation between Total Sales and Odometer (KM):", sales_odometer_corr)
print("Pearson Correlation between Price and Odometer (KM):", price_odometer_corr)
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

### Using the data correlation plotted above, we can see a correlation from total sales and odometer, and total sales and price. We do not see the same correlation from price and odometer as they have little to no correlation.
